```python
from keras.applications import InceptionV3
```

```python
inc=InceptionV3(input_shape=(224,224,3), weights= 'imagenet', include_top=False)
```

```python
for i in inc.layers:
  i.trainable = False
```

```python
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten


```

```python
x=Flatten()(inc.output)
```

```python
pred=Dense(10, activation='softmax')(x)
```

```python
from keras.models import Model
```

```python
model=Model(inputs=inc.input, outputs=pred)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=('accuracy'))
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# Define the input shape
input_shape = (32, 32, 3)
```

```python
# Define the model architecture
model = Sequential()
```

```python
# Add the first convolutional layer
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D((2, 2)))
```

```python
# Add a second convolutional layer
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
```

```python
# Add a third convolutional layer
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
```

```python
# Flatten the output from the convolutional layers
model.add(Flatten())
```

```python
# Add a fully connected layer with 512 units
model.add(Dense(512, activation='relu'))
```

```python
# Add a dropout layer to prevent overfitting
model.add(Dropout(0.5))
```

```python
# Add the output layer with 10 units (for a 10-class classification problem)
model.add(Dense(10, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Print the model summary
model.summary()
```
